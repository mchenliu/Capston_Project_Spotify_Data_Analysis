## Top 10 Genres

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# load the cleaned data
artist_genre_df = pd.read_csv('../Cleaned_Data/Completed_Artist_Genre_List.csv')

# Split the 'genres' column into individual genres
all_genres = artist_genre_df['genres'].dropna().str.split(', ').explode()

# count genre occurances
genre_count = all_genres.value_counts()
# identify top 10 genres
top_genre = genre_count.head(10)
```

### Visualize Top 10 Genres

```python
# select top 10 genres and sort in descending order
genre_count_sorted = genre_count.sort_values(ascending=False).head(10)


# create pie chart with matplotlib
labels = top_genre.index
sizes=top_genre.values


plt.pie(
    sizes,
    labels=labels,
    #format percentage with 1 decimal place
    autopct='%1.1f%%',
    # rotate the first slice to start at 120 degree
    startangle=120,
    colors=sns.color_palette('Blues_r',n_colors=len(labels)),
    pctdistance=0.8 #pull percentage labels away from centre
)

plt.title('Genre Distribuion', fontsize=14)
plt.tight_layout()
plt.show()
```
